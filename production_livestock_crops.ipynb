```python
df_livestock = df_prod[df_prod['Element'] == 'Stocks']
```

```python
print('Categorical Values:')
cat_variables = df_livestock.select_dtypes(include=['object']).columns
for col in cat_variables:
    print(col)
    print(df_livestock[col].unique())
```

```python
excluded_areas = ['World','Africa', 'Eastern Africa', 'Middle Africa','Northern Africa','Southern Africa', 'Western Africa','Americas','Northern America','Central America'
,'Caribbean','South America' ,'Asia' ,'Central Asia' ,'Eastern Asia',
 'Southern Asia', 'South-eastern Asia', 'Western Asia', 'Europe',
 'Eastern Europe', 'Northern Europe', 'Southern Europe', 'Western Europe',
 'Oceania' ,'Australia and New Zealand', 'Melanesia', 'Micronesia',
 'Polynesia', 'European Union (27)', 'Least Developed Countries',
 'Land Locked Developing Countries', 'Small Island Developing States',
 'Low Income Food Deficit Countries',
 'Net Food Importing Developing Countries', 'China']
df_livestock = df_livestock[~df_livestock['Area'].isin(excluded_areas)]
```

```python
df_livestock = df_livestock[~df_livestock['Item'].str.contains('and')]
df_livestock_sorted = df_livestock.sort_values(by = 'Y2023', ascending = False)
df_livestock_sorted.head(10)
```
