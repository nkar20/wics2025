### INCOME

The income dataset has a bunch more races, so will have to combine or delete columns.
* Delete hispanic/latino column
* Delete multiple races
* There's a non-hispanic white and a white alone column, what would be the most accurate to the white column for the maternal death dataset? The "White Alone" column will be used rather than the "White excluding hispanic" because the data documentation for the maternal death dataset describes how death certificates might not have the accurate race recorded, especially for hispanic people.

```python
df_income.drop(columns = ['geoid', 'hisp_est', 'nhopi_est', 'api_est', 'nhwhite_est', 'other_est', 'othermore_est', 'twomore_est'], inplace = True)
```

```python
columns_to_drop = df_income.filter(like='_se').columns
df_income = df_income.drop(columns=columns_to_drop)
df_income.columns
```

```python
df_income.rename(columns={'name':'State',
                   'year':'Year',
                   'total_est':'Total',
                   'aian_est':'American Indian or Alaska Native',
                   'asian_est':'Asian or Pacific Islander',
                   'black_est':'Black or African American',
                   'white_est':'White'}, inplace = True)
df_income.columns
```

```python
df_income.head()
```

```python
df_income.describe()
```

```python
def fill_nan(df, race_columns, group_col):
    for race_col in race_columns:
        # Compute the mean of the race column by state
        group_mean = df.groupby(group_col)[race_col].transform('mean')
        # Replace NaN with the computed mean
        df[race_col] = df[race_col].fillna(group_mean)
    return df

# List of race columns
race_columns = ['American Indian or Alaska Native', 'Asian or Pacific Islander', 'Black or African American', 'White']

# Apply the function
df_income = fill_nan(df_income, race_columns, 'State')

df_income
```

```python
print(df_income.isnull().sum())
```

```python
melted_df_income = df_income.melt(
    id_vars = ['State', 'Year'],  # Columns to keep as-is
    value_vars = ['American Indian or Alaska Native', 'Asian or Pacific Islander', 'Black or African American', 'White'],  # Columns to "melt"
    var_name = 'Race',  # New column to label the melted variables
    value_name = 'Income'  # New column to label the values
)
melted_df_income
```

```python
df_death_total = df_death.groupby(['Year', 'State', 'Race', 'Cause of Death', 'Maternal Deaths']).agg({'Population': 'sum'}).reset_index()
df_death_total[['Population']] = df_death_total[['Population']].astype('int')
df_death_total
```

```python
df_death['Crude Rate'] = df_death['Maternal Deaths'] / df_death['Female Population'] * 100000
```
